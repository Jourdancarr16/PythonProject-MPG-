```python
#1 - How do the car brand/name contribute to the fuel economy?
mpg.name.unique()
```

```python
import numpy as np
mpg['brand'] = np.where(mpg.name.str.contains("audi"), 'Audi'
                        ,np.where(mpg.name.str.contains("buick"), 'Buick'
                                  ,np.where(mpg.name.str.contains("chevy|chevrolet"), 'Chevy/Chevrolet'
                                            ,np.where(mpg.name.str.contains("dodge"), 'Dodge'
                                                      ,np.where(mpg.name.str.contains ("fiat"), 'Fiat'
                                                                ,np.where(mpg.name.str.contains("ford"), 'Ford'
                                                                          ,np.where(mpg.name.str.contains("honda"), 'Honda'
                                                                                    ,np.where(mpg.name.str.contains("mazda"), 'Mazda'
                                                                                              ,np.where(mpg.name.str.contains("toyota"), 'Toyota'
                                                                                                        ,np.where(mpg.name.str.contains("volkswagen"), 'Volkswagen'
                                                                                                                  ,np.where(mpg.name.str.contains("volvo"), 'Volvo', 'Other')))))))))))
```

```python
sns.catplot(data=mpg, x="mpg", y="brand", kind="bar", errorbar=None)
```

```python
#2 Brands by Origin & MPG
```

```python
sns.relplot(data=mpg, x='mpg', y='brand', col="origin")
```

```python
#Based off of the information above, the graphs show that the USA have the most brands of cars with MPG under 20.
#Europe had the most car brands with MPG over 40 MPG. Japan had the highest average of MPG compared to Europe and the US.
```
